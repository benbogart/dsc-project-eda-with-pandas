```python
# Let's get started importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the data
df = pd.read_csv('ames_train.csv')
```

```python
# Investigate the Data
df.info()
```

```python
df.Fence.value_counts
```

```python
df.describe()
```

* Use built-in Python functions to explore measures of centrality and dispersion for at least 3 variables


```python
print('Sale Price Mean', df['SalePrice'].mean())
print('Sale Price Standard Deviation', df['SalePrice'].std())
```

```python
#Sale prices by size of garage (in cars)
#df.groupby('GarageCars')['SalePrice'].plot(alpha = 0.3, subplots=True)
#plt.legend(title = "# Cars")

grouped = df.groupby('GarageCars')['SalePrice']
                      # fix up if odd number of groups
fig, axs = plt.subplots(figsize=(6,3), 
                        nrows=2, ncols=3,
                        gridspec_kw={'hspace':0.4}) # Much control of gridspec

targets = zip(grouped.groups.keys(), axs.flatten())
for i, (key, ax) in enumerate(targets):
    ax.hist(grouped.get_group(key), bins=25)
    ax.set_title('#cars: %d'%key)

fig.delaxes(axs[1, 2])
    
plt.show()
```

```python
#Lets compare the correlation of GrLivArea (Ablove Ground Living Area) and LotArea to Price for the years 2000 - 2010

df_decade = df.loc[(df['YrSold'] >= 2000) & (df['YrSold'] <= 2010), ['GrLivArea', 'LotArea', 'SalePrice']]
print('Correlation Coeficient LotArea to SalePrice',
      round(df_decade[['LotArea', 'SalePrice']].corr().loc['LotArea','SalePrice'], 2))
# df_decade[['GrLivArea', 'SalePrice']].corr()

print('Correlation Coeficient Ground Living Area to SalePrice',
      round(df_decade[['GrLivArea', 'SalePrice']].corr().loc['GrLivArea','SalePrice'], 2))

#Above ground living area is much more corelated with sale price than the size of the lot.
```
